## Test Section

This is a bit of basic code to generate a random graph and display it

```python
import matplotlib.pyplot as plt
import numpy as np
import networkx as nx
import pandas as pd

nodes = list(range(20))

df = pd.DataFrame({"from": np.random.choice(nodes, 20), "to": np.random.choice(nodes, 20)})

G = nx.from_pandas_edgelist(df, source="from", target="to")

nx.draw_planar(G)
plt.show()
```

## Scale Free Graph Generation


```python
SF = nx.barabasi_albert_graph(200,1)

pos = nx.kamada_kawai_layout(SF)
#pos = nx.spring_layout(SF, k=1.4, iterations=600)

nx.draw(SF, pos, node_size=50)
plt.show()
```

```python
# HIGHEST DEGREE

import random

seeds = 4

# find the minimum value of k for the number of seed nodes
k_min = min(sorted(dict(SF.degree()).values(), reverse=True)[:seeds])
    
# select all nodes with k above k_min
high_k_nodes = [n for n, k in dict(SF.degree()).items() if k > k_min]

# select all nodes with k = k_min
k_min_nodes = [n for n, k in dict(SF.degree()).items() if k == k_min]

# nodes we need randomly pick from k_min_nodes to meet seed size
sample_size = seeds - len(high_k_nodes)

# select seed nodes and update seed statistics
seed_nodes_ = high_k_nodes + random.sample(k_min_nodes, sample_size)


```
